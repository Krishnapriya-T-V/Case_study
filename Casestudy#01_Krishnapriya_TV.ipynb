# 4. Find the top 5 cars based on their sales price.

```python
data2=data1.sort_values(by=['Sale Price'], ascending=False)
data3=data2.head(5)
data3
```

```python
sns.scatterplot(data=data3, x="Make", y="Sale Price")
```

```python
##Insights:Volkswagen Jetta III,Audi A6,Lexus LX,Mazda MX-6 are the five top cars based on their seles price.The range of seles price lies between 99970-100000.Lexus has two models(LX and ES)comes n top 5. 
```
